```python
import pandas as pd

yr = [2001,2002,2005,2010,2018,2020]

f1= {2001:20,2002:12,2005:15,2010:17,2018:105,2020:200}
fir1=pd.Series(f1, index=yr)

f2= {2001:40,2002:2,2005:8,2010:30,2018:210,2020:250}
fir2=pd.Series(f2, index=yr)

f3= {2001:48,2002:52,2005:68,2010:87,2018:178,2020:253}
fir3=pd.Series(f3, index=yr)

df3=pd.DataFrame(columns=["CSE", "EEE", "TEXTILE"], index=yr)

df3.CSE= fir1
df3.EEE= fir2
df3.TEXTILE= fir3

df3
```
